```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Profit from merchandize and food stalls:
    # Constraints: Number of cashiers
    # Appropriate function, that captures diminishing returns: Logistic Function
    # This can be further optimized from using the data obtained from the previous sales (future scope) or may be modelled completely differently according to the situation

# Profit for Main stage
    # Incentivize more number of people, hence quadratic

#----

# Profit Models

class logistic_profit:
    def __init__(self, half_sat=500, capacity=1000, unit_profit=40):
        self.half_sat = half_sat
        self.capacity = capacity
        self.unit_profit = unit_profit
        
    def profit(self, num_people):
        return self.capacity * self.unit_profit * num_people / (self.half_sat + num_people)
    
    def sat_profit(self):
        return self.capacity * self.unit_profit

class quadratic_profit:
    def __init__(self, linear = 10, quadratic = 0.01):
        self.linear = linear
        self.quadratic = quadratic
        
    def profit(self, num_people):
        return self.linear * num_people + self.quadratic * num_people**2

merch_model = logistic_profit()
food_model = logistic_profit()
stage_model = quadratic_profit()

merch_model.half_sat = 500
merch_model.capacity = 1000
merch_model.unit_profit = 40

food_model.half_sat = 1250
food_model.capacity = 3000
food_model.unit_profit = 20

stage_model.linear = 10
stage_model.quadratic = 0.01

# Plotting the profits

x = np.linspace(0, 2500, 10000)

y1 = merch_model.profit(x)
y2 = food_model.profit(x)
y3 = stage_model.profit(x)

plt.grid()
plt.plot(x, y1, label='Merch Profit', color='blue')
plt.plot(x, y2, label='Food Profit', color='green')
plt.plot(x, y3, label='Main Stage Profit', color='red')
plt.xlabel("Attendees")
plt.ylabel("Expected Profit")
plt.title("Logistic Function Profit Model")
plt.legend()
plt.show()

# print("Saturation Profit for Merch: ", merch_model.sat_profit())
# print("Saturation Profit for Food: ", food_model.sat_profit())
```
